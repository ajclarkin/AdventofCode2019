```python
def GetInput():
    '''
        Does what it says on the tin - reads the input file.
    '''
    data = [int(x) for x in open('input.txt').read() if x != '\n']
    return data



def RowBaseArray(rowno):
    '''
        Given a rowno return the base pattern unrepeated for use in testing.
        Do not strip off the leading 0.
        The first row = 1.
    '''

    base = [0, 1, 0, -1]
    rowbase = [b for b in base for i in range(rowno)]
    return rowbase
```

```python
def TestingMultiplySingleRow(input):
    '''
        Given an input it generates the first few rows worth of data.
        The input is 650 chars long. This gives the sum of each 650 char group.
        We can see a repeating pattern:
            For row number n, the first groups give values, the second n groups give the negated values, and then it repeats.
        We can use this pattern to shorten the processing.
    '''
    length = len(input)


    for rowno in range(1, 30):
        b = 1

        base = RowBaseArray(rowno)

        print(f'\nRow: {rowno}', end="\t")
        for j in range(40):

            row_sum = 0
            
            for i in range(length):
                row_sum = row_sum + (input[i] * base[b])
                # print(f'{b} - {base[b]}')
                b = b + 1 if b < len(base)-1 else 0

            # print(f'Row sum: {row_sum}')
            print(f'{row_sum},', end = ' ')
            # print(f'New iteration')

            # input = answer.copy()
            # Row value becomes the (positive) rightmost digit of row_sum
            # row_value = int(str(row_sum)[-1])
            # answer.append(row_value)
```
